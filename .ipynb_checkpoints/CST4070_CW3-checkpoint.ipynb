```python
import pandas as pd
import matplotlib.pyplot as plt

global_mobility_report = pd.read_csv("../Global_Mobility_Report_sample.csv", sep=",")
owid_covid_data = pd.read_csv("../owid-covid-data.csv", sep=",")

global_mobility_report['date'] = pd.to_datetime(global_mobility_report['date'])
owid_covid_data['date'] = pd.to_datetime(owid_covid_data['date'])

owid_covid_data = owid_covid_data[(owid_covid_data.continent.eq("Europe")) & (owid_covid_data['date'].dt.weekday >3)]
european_country = pd.unique(owid_covid_data.location)

global_mobility_report = global_mobility_report[(global_mobility_report.country_region.isin(european_country)) & (global_mobility_report['date'].dt.weekday >3)]

df = pd.merge(left=global_mobility_report, right=owid_covid_data, left_on=['date', 'country_region'], right_on=['date', 'location'])
df = df.drop(columns = ['country_region_code', 'sub_region_1', 'sub_region_2', 'iso_3166_2_code', 'metro_area', 'census_fips_code', 'continent', 'total_vaccinations_per_hundred', 'extreme_poverty', 'icu_patients', 'icu_patients_per_million', 'weekly_icu_admissions', 'reproduction_rate', 'total_vaccinations', 'country_region', 'aged_70_older', 'aged_65_older', 'stringency_index', 'human_development_index', 'life_expectancy', 'handwashing_facilities', 'new_vaccinations_per_million', 'new_vaccinations', 'female_smokers', 'male_smokers', 'cardiovasc_death_rate', 'diabetes_prevalence', 'population', 'hospital_beds_per_thousand', 'population_density', 'median_age', 'iso_code', 'hosp_patients', 'weekly_hosp_admissions_per_million', 'weekly_icu_admissions_per_million', 'weekly_hosp_admissions', 'hosp_patients_per_million', 'positive_rate', 'tests_per_case', 'tests_units', 'gdp_per_capita'])

df = df.dropna()

df
```

```python
df_corr = df.corr()
df_corr
```

```python
import seaborn as sn

sn.heatmap(df_corr)
plt.show()
```

```python
graph1 = df[['date', 'retail_and_recreation_percent_change_from_baseline']]
graph1.groupby('date').mean().plot(kind='line')
plt.figure(figsize=(32, 16))
plt.ylim([-100,20])
plt.grid(b=True, which='major', color='g', linestyle='--')
```

```python
graph2 = df[['date', 'retail_and_recreation_percent_change_from_baseline', 'location']]
plt.figure(figsize=(32, 16))
graph2.groupby('location').plot.line(x='date', y='retail_and_recreation_percent_change_from_baseline')
plt.legend()
```
